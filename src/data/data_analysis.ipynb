# Data Analysis

We decided it would be most beneficial to us to analyse the data and check which variables would be relevant to our model. To start, we imported our data into Pandas Dataframes.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
conn = sqlite3.connect('database.db')

factory = pd.read_sql_query('SELECT * FROM factory', conn)
system = pd.read_sql_query('SELECT * FROM system', conn)
expansion_model = pd.read_sql_query('SELECT * FROM expansion_model', conn)
sensor_model = pd.read_sql_query('SELECT * FROM sensor_model', conn)
system_model = pd.read_sql_query('SELECT * FROM system_model', conn)
expansion = pd.read_sql_query('SELECT * FROM expansion', conn)
sensor = pd.read_sql_query('SELECT * FROM sensor', conn)

conn.close()


print("factory")
print(factory.head(5))
print("====================================================================")
print("sensor_model")
print(sensor_model.head(5))
print("====================================================================")
print("system")
print(system.head(5))
print("====================================================================")
print("expansion")
print(expansion.head(5))
print("====================================================================")
print("expansion_model")
print(expansion_model.head(5))
print("====================================================================")
print("system_model")
print(system_model.head(5))
print("====================================================================")
print("sensor")
print(sensor.head(5))
```

Let's check the number of sensors that are currently operating.

```python
status_counts = sensor['status'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Status Types')
plt.show()
```

Let's see if there are any sensors that differ greatly from the average.

```python
sns.boxplot(data=sensor_model, x='min_value')
plt.show()
```

```python
sns.boxplot(data=sensor_model, x='max_value')
plt.show()
```

We'll see if there is any correlation between the minimum and maximum values of the sensor.

```python
merged_sensor = sensor.merge(sensor_model[['id', 'min_value', 'max_value']], left_on='sensor_model_id', right_on='id', how='left')
merged_sensor['status_encoded'] = pd.factorize(merged_sensor['status'])[0]
merged_sensor.pop('status')
print("0 - Maintenance required")
print("1 - Working")
print("2 - Online")
print("3 - Offline")
merged_sensor = merged_sensor[['status_encoded', 'min_value','max_value']]
corr = merged_sensor.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
