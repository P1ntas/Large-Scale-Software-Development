Imported libraries and defined constants.

```python
import math
pi = math.pi
import csv
import matplotlib.pyplot as plt
```

# FLOW RATE

Code to generate reference data for the flow rate sensor.

# Temperature

Code to generate reference data for the temperature sensor.
- MinTemperature = 30
- MaxTemperature = 90
- FullRefreshTime = 40
- FunctioningTimeWithoutRefresh = 50
- FunctioningTimeWithRefresh = 90

```python
# - - - - - - - - - - - - - - - - - - - -
# REFERENCE DATA
# - - - - - - - - - - - - - - - - - - - -

temperatures = []

# Time 0 to 10
for i in range(0, 10):
    temperatures.append(30)

# Time 10 to 40
for i in range(0, 30):
    t = 30*math.sin((pi/30)*i-(pi/2))+60
    temperatures.append(t)

# Time 40 to 50
for i in range(0, 10):
    temperatures.append(90)

# Refresh time (40, Time 50 to 90)
for i in range(0, 40):
    t = (3/80)*((i-40)**2)+30
    temperatures.append(t)

# - - - - - - - - - - - - - - - - - - - -
# CSV
# - - - - - - - - - - - - - - - - - - - -

# Specify the file path
file_path = 'temperature_reference.csv'

# Write data to CSV
with open(file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Time', 'Temperature'])  # Write header
    for time, temperature in enumerate(temperatures):
        writer.writerow([time, temperature])

# - - - - - - - - - - - - - - - - - - - -
# PLOT
# - - - - - - - - - - - - - - - - - - - -

plt.plot(temperatures)
plt.title("Reference Temperature")
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()
```

# Pressure

Code to generate reference data for the pressure sensor.
- MinPressure = 50
- MaxPressure = 380
- FullResetTime = 60
- FunctioningTimeWithoutReset = 57
- FunctioningTimeWithReset = 117

```python
# - - - - - - - - - - - - - - - - - - - -
# REFERENCE DATA
# - - - - - - - - - - - - - - - - - - - -

pressures = []

# Time 0 to 5
for i in range(0, 5):
    pressures.append(50)

# Time 5 to 20
for i in range(0, 15):
    t = 85*math.sin((pi/15)*i-(pi/2))+135
    pressures.append(t)

# Time 20 to 30
for i in range(0, 10):
    pressures.append(220)

# Time 30 to 37
for i in range(0, 7):
    t = 7.5*math.sin((pi/7)*i-(pi/2))+227.5
    pressures.append(t)

# Time 37 to 45
for i in range(0, 8):
    pressures.append(235)

# Time 45 to 53
for i in range(0, 8):
    t = 72.5*math.sin((pi/8)*i-(pi/2))+307.5
    pressures.append(t)

# Time 53 to 57
for i in range(0, 4):
    pressures.append(380)

# Reset time (55, Time 57 to 118)
for i in range(0, 61):
    t = (5.5/60)*((i-60)**2)+50
    pressures.append(t)

# - - - - - - - - - - - - - - - - - - - -
# CSV
# - - - - - - - - - - - - - - - - - - - -

# Specify the file path
file_path = 'pressure_reference.csv'

# Write data to CSV
with open(file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Time', 'Pressure'])  # Write header
    for time, pressure in enumerate(pressures):
        writer.writerow([time, pressure])

# - - - - - - - - - - - - - - - - - - - -
# PLOT
# - - - - - - - - - - - - - - - - - - - -

plt.plot(pressures)
plt.title("Reference Pressure")
plt.xlabel("Time")
plt.ylabel("Pressure")
plt.grid(True)
plt.show()
```
